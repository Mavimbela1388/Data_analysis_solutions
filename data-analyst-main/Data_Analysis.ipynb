```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# data for andriod games for 2017
df= pd.read_csv('platform1_data.csv')
df
```

```python
# andriod games conributions from 01 August 2017 
df2= pd.read_csv('all_contributions.csv')
df2
```

```python
df3= pd.read_csv('yearly_contributions.csv')
df3
```

```python
x= df3['day']
x1=df2['day']
```

```python
y1=df3['Game_Time_Contribution']
y2 = df3['Game_Microtrans_Contribution']
y3= df3['Game_Accounts_contribution']

y4=df2['Game_Time_Contribution']
y5 = df2['Game_Microtrans_Contribution']
y6= df2['Game_Accounts_contribution']
```

```python
plt.subplot(1,3,1)
fig1=plt.scatter(x,y1, c= 'C76')
fig2=plt.scatter(x1,y4, c= 'C6')
plt.autoscale(enable= True,axis= 'y1', tight= True )
plt.title('Time spent')
plt.xlabel('Day')
plt.ylabel('Game Time Contribution in %')




plt.subplot(1,3,2)
plt.scatter(x,y2, c= 'C76')
plt.scatter(x1,y5, c= 'C89')
plt.autoscale(enable= True,axis= 'y2', tight=False )
plt.title('Game Accounts')
plt.xlabel('Day')
plt.ylabel('Game Accounts contribution in  %')


plt.subplot(1,3,3) 
plt.scatter(x,y3, c= 'C36')
plt.scatter(x1,y6, c= 'C68')
plt.autoscale(enable= True,axis= 'y1', tight= False)
plt.title('Microtransactions')
plt.xlabel('Day')
plt.ylabel('Game_Time_Contribution in %')
plt.show
```

```python
## from the figures above:
#the percentage time spent playing the games on andriod  mobile device harply increased the first day of introducing the stategy, this can be seen by the blue scatter points 
#the percentage of unique game accounts created also sharply increased the first day of introducing the strategy,
#and the contribution of the microtransactions made by games on andriod device increased the first day of introducing the stragety,
# in all the three graphs we can see that the percentage contribution of time , game accounts and money made from microtransactions flactuate 
# 
```
